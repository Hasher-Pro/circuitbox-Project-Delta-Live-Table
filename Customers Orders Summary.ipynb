# Create Customers Order Summary
1. Join the table silver_customers, silver_addresses, and silver_orders
2. Retrieve the latest addresses of the customers
3. Calculate the following
- total_order
- total_items_ordered
- total_orders_amount

![image_1771858254579.png](./image_1771858254579.png "image_1771858254579.png")

```sql
create or refresh materialized view gold_customer_order_summary
as
select
c.customer_id,
c.customer_name,
c.date_of_birth,
c.telephone,
c.email,
a.address_line_1,
a.city,
a.state,
a.postcode,
count(Distinct o.order_id) as total_orders,
sum(o.item_quantity) as total_item_quantity,
sum(o.item_quantity * o.item_price) as total_order_amount
from LIVE.silver_customers c
join live.silver_addresses a
on c.customer_id = a.customer_id
join live.silver_orders o
on c.customer_id = o.customer_id
where a.__END_AT is null
group by all;
```
