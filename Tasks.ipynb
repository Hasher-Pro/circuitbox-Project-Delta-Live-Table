```python
# Dog Years Converter

human_years = float(input("Enter the number of human years: "))

# Check for negative input
if human_years < 0:
    print("Error: Age cannot be negative.")
else:
    # If age is less than 2 years
    if human_years <= 2:
        dog_years = human_years * 10.5
    else:
        dog_years = 2 * 10.5 + (human_years - 2) * 4

    print("That is equivalent to", dog_years, "dog years.")
```

```python
import math

# Get input from user
a = float(input("Enter value of a: "))
b = float(input("Enter value of b: "))
c = float(input("Enter value of c: "))

# Ensure a is not zero
if a == 0:
    print("Error: 'a' cannot be zero in a quadratic equation.")
else:
    # Calculate discriminant
    discriminant = b**2 - 4*a*c

    # Case 1: No real roots
    if discriminant < 0:
        print("The equation has no real roots.")

    # Case 2: One real root
    elif discriminant == 0:
        root = -b / (2*a)
        print("The equation has one real root:")
        print("Root =", root)

    # Case 3: Two real roots
    else:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        print("The equation has two real roots:")
        print("Root 1 =", root1)
        print("Root 2 =", root2)
```

```python
def edit_distance(s, t):
    # Base cases: If one string is empty, the distance is the length of the other
    if len(s) == 0:
        return len(t)
    if len(t) == 0:
        return len(s)

    # Check last characters
    if s[-1] == t[-1]:
        cost = 0
    else:
        cost = 1

    # Recursive cases
    d1 = edit_distance(s[:-1], t) + 1      # Deletion
    d2 = edit_distance(s, t[:-1]) + 1      # Insertion
    d3 = edit_distance(s[:-1], t[:-1]) + cost  # Substitution

    return min(d1, d2, d3)


# Main program
string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

distance = edit_distance(string1, string2)

print("The edit distance between the strings is:", distance)
```

```python
def edit_distance(s, t):
    """
    Computes the Levenshtein edit distance between strings s and t
    using the recursive algorithm provided.
    """
    # Base Case: If one string is empty, the distance is the length of the other
    if len(s) == 0:
        return len(t)
    elif len(t) == 0:
        return len(s)
    
    else:
        # Check if the last characters match
        cost = 0
        if s[-1] != t[-1]:
            cost = 1
        
        # d1: Deletion from s (or insertion into t)
        d1 = edit_distance(s[:-1], t) + 1
        
        # d2: Insertion into s (or deletion from t)
        d2 = edit_distance(s, t[:-1]) + 1
        
        # d3: Substitution (or no change if cost is 0)
        d3 = edit_distance(s[:-1], t[:-1]) + cost
        
        # Return the path with the minimum number of operations
        return min(d1, d2, d3)

def main():
    print("--- String Edit Distance Calculator ---")
    string1 = input("Enter the first string: ")
    string2 = input("Enter the second string: ")
    
    distance = edit_distance(string1, string2)
    
    print(f"\nThe edit distance between '{string1}' and '{string2}' is {distance}.")

if __name__ == "__main__":
    main()
```

```python
class Triangle:
    # Class variable: every triangle instance will have 3 sides
    number_of_sides = 3

    def __init__(self, angle1, angle2, angle3):
        """
        Constructor to initialize the angles of the triangle.
        """
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3

    def check_angles(self):
        #Method to verify if the sum of angles equals 180 degrees.
        total_sum = self.angle1 + self.angle2 + self.angle3
        if total_sum == 180:
            return True
        else:
            return False

# Creating an instance of the class
my_triangle = Triangle(120, 30, 60)

# Printing the required values
print(f"Number of sides: {my_triangle.number_of_sides}")
print(f"Are the angles valid? {my_triangle.check_angles()}")
```
