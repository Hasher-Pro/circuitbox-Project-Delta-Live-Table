### Create External Location 
- Make sure to verify the location using %fs ls 

```python
%sql
SHOW CATALOGS
```

```python
%sql
create external location if not exists circuitbox_ext
url 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/'
with (storage credential `circuitbox-storage-cred`)
comment 'External location for circuitbox'
```

```python
%fs ls 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/'
```

### Create Catalog
- using managed location

```python
%sql
create catalog if not exists circuitbox
managed location 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/'
comment 'Catalog for circuitbox'
```

```python
%sql
SHOW CATALOGS
```

### Create Schema for circuitbox
- Landing and Lakehouse schema

```python
%sql
use catalog circuitbox;
create schema if not exists landing
managed location 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/landing';

create schema if not exists lakehouse
managed location 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/lakehouse';
```

### Create External Volumne operational location 
- using the created catalog and schema 

```python
%sql
use catalog circuitbox;
use schema landing;

create external volume if not exists operational_data
location 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/landing/operational_data'
```

```python
%fs ls 'abfss://circuitbox@databrickstorageacc.dfs.core.windows.net/landing/operational_data/'
```
